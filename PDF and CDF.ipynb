## Calculate PDF and CDF

```python
counts,bins  = np.histogram(iris_setosa.PL,bins=10)
```

```python
bins # bins will always be 1 more than given since it gives edgegs and dhistogram calculates based on range
```

```python
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
```

```python
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)
```

### How to use CDF in out analysis

```python
plt.style.use('fivethirtyeight')

counts,bins  = np.histogram(iris_setosa.PL,bins=10)

pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)

counts,bins  = np.histogram(iris_versicolor.PL,bins=10)

pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)

counts,bins  = np.histogram(iris_virginica.PL,bins=10)

pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)
```

### Analysis

1. If petal length is less than 2 then 100% of the flowers are setosa.
2. If petal length is greater than 2 and less than 5 then it is versicolor with probability pf 95%.
3. If petel length is greater than 5 then it is verginica but we can be wrong 10% since for 10% data it is less than 5.
